```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sodapy import Socrata
cliente=Socrata ('www.datos.gov.co', None )
result=cliente.get('e7w8-zhsx')
df_n=pd.DataFrame.from_records(result)
df_n.head()
```

```python
df_n.info()
```

```python
df_n.describe()
```

```python
#para ver todas las columnas
pd.options.display.max_columns=None
df_n
```

```python
Pesogr=pd.to_numeric(df_n['peso_gramos'],errors='coerce')
Estatura=pd.to_numeric(df_n['talla_cent_metros'],errors='coerce')
T_gestacion=pd.to_numeric(df_n['tiempo_de_gestaci_n'],errors='coerce')
df_n['Pesogr']=Pesogr
df_n['Estatura']=Estatura
df_n['T_gestacion']=T_gestacion
df_n

```

```python
float(Estatura[0])
```

```python
#Clasificacion
#La idea es clasificar a los nacidos por su peso, para saber si nacieron con peso Bajo, adecuado o elevado
new_df_n=pd.DataFrame(df_n['Pesogr']/1000)
new_df_n['Estatura_metros']=df_n['Estatura']/100
new_df_n['IMC']=new_df_n['Pesogr']/(new_df_n['Estatura_metros']**2)
Rangos=[0,10,90,]
Valores=[1,2,3]
new_df_n['Codigo']=np.digitize(new_df_n['IMC'], Rangos, right=True)
new_df_n['Rango_codificado']= new_df_n['Codigo'].map({1: 'IMC bajo ', 2: 'IMC Adecuado', 3: 'IMC Alto'})
new_df_n
```

```python
new_df_n.info()
```

```python
Salida=pd.to_numeric(new_df_n['Codigo'], errors='coerce')
print(type(Salida))
y=np.array(Salida)
print(type(y))
```

```python
Codigo=new_df_n['Codigo']

x1=np.array(Codigo)
Salida=[]
for i in range(len(x1)):
  aux=[]
  aux.append(x1[i])
  Salida.append(aux)
X=np.array(Salida)
X

```

```python
print(X[0])
print('---------')
print(y[0])
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y, test_size=0.15)
print(X_test)
print('--------------------------')
print(y_test)
```

```python
#Aplicar el algoritmo de clasificación de los K vecinos más cercanos
#Knn=KNeighborsClassifier(n_neighbors=3)
#Knn.fit(X_train,y_train)
from sklearn.cluster import KMeans
Kmeans=KMeans(n_clusters=2, max_iter=300)
Kmeans.fit(X_train,y_train)
```

```python

from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor(n_neighbors=3)
knn.fit(X_train, y_train)
```

```python
#Calcular los puntajes de la clasificación
objetivos=Kmeans.fit_predict(X_test,y_test)
objetivos
```

```python
Kmeans.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
cm= confusion_matrix(y_test,knn.predict(X_test))
cm
```

```python
import seaborn as sns
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, cmap="coolwarm")
plt.xlabel('Valor x')
plt.ylabel('Valor y')
plt.title('Matriz de Confusión')
plt.show()
```

**Conclusion **

En ese metodo vemos que el modelo de clasificacion que usamos ha logrado una puntuacion perfecta en los datos de prueba, -0.0 indica que los puntos de los datos estan muy cerca de los centroides asignados en el modelo.

Por otro laado al realizar la matriz de confusion, nos dimos cuenta que tuvimos 2 muestras de la clase IMC baja   que fueros clasificados correctamente  y 148 muestras de la clase IMC Adecuada  que fueron clasificados correstamente,esto demuestra que no tuvimos clasificaciones incorrectas.

```python
def contar(n, vec):
  '''Para esta funcion n es el numero a contar y vec es la lista donde contar '''
  count=0
  for i in vec:
    if i==n:
      count+=1
  return count
```
